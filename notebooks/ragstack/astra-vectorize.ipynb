```python
import getpass
astra_token = getpass.getpass("Astra token:")
astra_endpoint = getpass.getpass("Astra db endpoint:")
openai_key = getpass.getpass("OpenAI Key:")
collection = input("Collection name:")

from langchain.vectorstores.astradb import AstraDB 
from langchain.embeddings import OpenAIEmbeddings

embedding = OpenAIEmbeddings(openai_api_key=openai_key)
vector_db = AstraDB(
        collection_name=collection,
        embedding=embedding,
        token=astra_token,
        api_endpoint=astra_endpoint
    )
print("Astra configured")
```

```python
import csv
import requests
import io

texts = []
metadatas = []
ids = []

url = 'https://raw.githubusercontent.com/nicoloboschi/awesome-rag/main/recommendations/recommendations/products.csv'
r = requests.get(url)
csv_reader = csv.DictReader(io.StringIO(r.text), delimiter=';', quotechar='"')
for i, row in enumerate(csv_reader):
    texts.append(row["DESCRIPTION"])
    metadatas.append(row)
    ids.append(i)
vector_db.add_texts(texts, metadatas, ids=ids)
```

```python
print(vector_db.astra_db.collection(collection).find())
```
